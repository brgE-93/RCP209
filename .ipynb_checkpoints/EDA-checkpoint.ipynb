# Statlog Heart

 ## Analyse et exploration des donees :

```python
import pandas as pd 
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
Heart=pd.read_csv('ProjectRCP/heart.dat',sep=" ",header = None)
Heart.columns = ["age","sex","chest pain type","resting blood pressure","serum cholestoral in mg/dl","fasting blood sugar > 120 mg/dl","resting electrocardiographic results  (values 0,1,2)",
               "maximum heart rate achieved","exercise induced angina","oldpeak","the slope of the peak exercise ST segment","number of major vessels (0-3) colored by flourosopy","thal 3 = normal 6 = fixed defect 7 = reversable defect","Absence(1)/Presence(2) HeartDesease"]
```

```python
Data=Heart.copy()
```

```python
Data.head()
```

```python
print(Data.dtypes.value_counts())
Data.isna().sum()
```

### Visualisation de nos variables:

```python
#visualiser proportion de notre Target :
target=Data["Absence(1)/Presence(2) HeartDesease"]
print(target.value_counts()/Data.shape[0])
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
k=0
col=['resting blood pressure','age','serum cholestoral in mg/dl','maximum heart rate achieved','oldpeak']
for i in Data.columns[Data.columns.isin(col)]:
    plt.figure(figsize=(4,4))
    plt.subplot(1,1,k+1)
    sb.distplot(Data[i])
 
plt.show()
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
k=0
col=['resting blood pressure','age','serum cholestoral in mg/dl','maximum heart rate achieved','oldpeak','Absence(1)/Presence(2) HeartDesease']
for i in Data.columns[~Data.columns.isin(col)]:
    plt.figure(figsize=(4,4))
    plt.subplot(1,1,k+1)
    print((Data[i].value_counts().plot.pie(cmap=plt.get_cmap('YlGnBu'))))
    
plt.show()
```

  ### relation entre les variables/variable cible:

```python
target=Data["Absence(1)/Presence(2) HeartDesease"]
```

```python
#deux ensembles de dataframe selon target 
Absent_disease=Data[Data['Absence(1)/Presence(2) HeartDesease']==1]
Present_disease=Data[Data['Absence(1)/Presence(2) HeartDesease']==2]
```

```python
#variables categorielle: binaire,nominal ,ordinal
var_category=[Data.columns[1],Data.columns[2],Data.columns[5],Data.columns[6],Data.columns[8],Data.columns[10],
              Data.columns[11],Data.columns[12]]
for i in var_category:
    plt.figure()
    (sb.heatmap(pd.crosstab(Data[i],target),annot=True,fmt='d',cmap=plt.get_cmap('YlGnBu')))
plt.show()
```

```python
#variable continues/target
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
k=0
col=[Data.columns[0],Data.columns[3],Data.columns[4],Data.columns[7],Data.columns[9]]
for i in col :
    plt.figure(figsize=(4,4))
    plt.subplot(1,1,k+1)
    sb.distplot(Absent_disease[i],label='Absent')
    sb.distplot(Present_disease[i],label='Present')
    plt.legend()
```

### Analyser les correlations des variables explicatives
