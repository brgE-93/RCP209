```python
import pandas as pd 
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
Heart=pd.read_csv('heart.dat',sep=" ",header = None)
Heart.columns = ["age","sex","chest pain type","resting blood pressure","serum cholestoral in mg/dl","fasting blood sugar > 120 mg/dl","resting electrocardiographic results  (values 0,1,2)",
               "maximum heart rate achieved","exercise induced angina","oldpeak","the slope of the peak exercise ST segment","number of major vessels (0-3) colored by flourosopy","thal 3 = normal 6 = fixed defect 7 = reversable defect","Absence(1)/Presence(2) HeartDesease"]
```

# Preprocessing:

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import RobustScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_classif
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
#Nous allons effectues notre analyse sur une copie de notre dataframe
Df=Heart.copy()
```

```python
Df.columns = Df.columns.str.replace(' ','_')
```

```python
Df.head()
```

```python
X=Df.drop('Absence(1)/Presence(2)_HeartDesease',axis=1)
y=Df['Absence(1)/Presence(2)_HeartDesease']
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
print("Trainset shape",X_train.shape,y_train.shape,"Testset shape",X_test.shape,y_test.shape)
print(y_train.value_counts(),y_test.value_counts())
```

# Entrainement et optimisation :

```python
from sklearn import tree
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
```

```python
Svm_Model=make_pipeline(StandardScaler(),PCA(),SVC(random_state=0))
Adaboost_model=make_pipeline(SelectKBest(k=13),AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=1,random_state=0),random_state=0))
Tree_model=make_pipeline(SelectKBest(k=13),tree.DecisionTreeClassifier(random_state=0))
Random_forest_model=make_pipeline(SelectKBest(k=13),RandomForestClassifier( oob_score=True,random_state=0))
Regression_model=make_pipeline(SelectKBest(k=13),StandardScaler(),PCA(),LogisticRegression(max_iter=200,random_state=0))
```

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.model_selection import cross_validate
def Complete_report(model):
    scoring = {'accuracy' : make_scorer(accuracy_score), 
               'precision' : make_scorer(precision_score,pos_label=2),
               'recall' : make_scorer(recall_score,pos_label=2), 
               'f1_score' : make_scorer(f1_score,pos_label=2)
            }
    
    kfold= StratifiedKFold(n_splits=4,shuffle=False,random_state=None) 
    results =cross_validate(model,X_train,y_train,cv=kfold,scoring=scoring)
    print("f1 mean/f1 std :",np.mean(results.get('test_f1_score')),np.std(results.get('test_f1_score')))
    print("accuracy mean/accuracy std :",np.mean(results.get('test_accuracy')),np.std(results.get('test_accuracy')))
    print("precision mean/precision std :",np.mean(results.get('test_precision')),np.std(results.get('test_precision')))
    print("recall mean/recall std :",np.mean(results.get('test_recall')),np.std(results.get('test_recall')))
    return results
```

```python
#evaluation plus complete de nos modeles:
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

from sklearn.metrics import classification_report
from sklearn.model_selection import learning_curve
from sklearn.model_selection import StratifiedKFold

    
def Isoverfitting(model):
    t=np.linspace(0.1,1,10)
    c=make_scorer(f1_score,pos_label=2)
    N,train_score,val_score=learning_curve(model,X_train,y_train,scoring=c,train_sizes=t,cv=4)
    train_score_mean=train_score.mean(axis=1)
    val_score_mean=val_score.mean(axis=1)
    plt.plot(N,train_score_mean,label="train")
    plt.plot(N,val_score_mean,label="validationset")
    plt.show()
    
def matrice(model):
    predicted_targets =[]
    actual_targets =[]
    skf = StratifiedKFold(n_splits=4)
    for train_index, test_index in skf.split(X_train, y_train):
        Xx_train, Xx_test = X_train.take(list(train_index),axis=0), X_train.take(list(test_index),axis=0)
        yy_train, yy_test = y_train.take(list(train_index),axis=0), y_train.take(list(test_index),axis=0)
        model.fit(Xx_train,yy_train)
        predicted_targets=model.predict(Xx_test)
        cm=confusion_matrix(yy_test,predicted_targets)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        fig, ax= plt.subplots(figsize=(1,1))
        disp.plot(ax=ax)
        
    plt.show()
    
def Evaluate(model):
    model.fit(X_train,y_train)
    matrice(model)
    Complete_report(model)
    Isoverfitting(model)
```

```python
#etudier la variance de nos scores sur train et valset:
def Variance(model):
    N=100
    score_mean=[]
    for i in range(100):
        X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
        m= model
        m.fit(X_train, y_train)
        score_mean.append(cross_val_score(m,X_train,y_train,cv=4,scoring="accuracy").mean())
    return(np.std(score_mean))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from scipy.stats import uniform
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
import scipy
import seaborn as sns
```

## RandomizedSearchCv pour un seul model a la fois:

```python
def optimize_model(model,dico_params,scoring):
            dico_params=dico_params
            n_iter_search = 50
            grid=RandomizedSearchCV(model,param_distributions=dico_params,cv=4,scoring=scoring,n_iter=n_iter_search,random_state=0)
            %time grid.fit(X_train,y_train)
            bestparams=grid.best_params_
            bestscore=grid.best_score_
            new_model=grid.best_estimator_

            print(f'{"bestparams":-<30} {bestparams}')
            print(f'{"bestscore":-<30}{bestscore}')
            return new_model
```

### courbe precision-recall 

```python
from sklearn.metrics import precision_recall_curve
```

```python
#fonction de transformation:
def binary_labels(y):
    code={1:0,
         2:1}
    y_temp=y.map(code)
    return y_temp
```

```python
#observer futur precision /sensibilite
precision,recall,threshold=precision_recall_curve(binary_labels(y_test),new_model_LR.decision_function(X_test))
plt.show()
plt.title("PR Curve for logistic regression")
plt.plot(threshold,precision[:-1],label='precision')
plt.plot(threshold,recall[:-1],label='recall')
plt.legend()
```

```python
def model_final(model,X,threshold=0):
    zz=model.decision_function(X)> threshold
    return zz
```

```python
y_pred_LR=model_final(new_model_LR,X_test,threshold=-0.245)
```

```python
f1_score(binary_labels(y_test),y_pred_LR)
```

```python
print(recall_score(binary_labels(y_test),y_pred_LR))
print(precision_score(binary_labels(y_test),y_pred_LR))
```
